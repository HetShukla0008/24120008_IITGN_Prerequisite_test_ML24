# Q1


Total multiplications = n x m x k
Total additions = n x m x (k-1)

In matrix multiplication of (n x k) and (k x m) matrices, each element of each row of matrix-1 is multiplied with each corresponding element of each column of matrix-2. Thus for multiplication of a single row having k elements with single column having k elements has total k multiplications.These k multiplications are added to form a single element in product matrix so there are corresponding k-1 additions here. Now there are m columns in matrix-2, so each row is further multiplied m times, each having k multiplications and k-1 additions. So the number of multiplications and additions for each row of matrix-1 is m x (k) and m x (k-1) respectively. Further, there are n rows so total multiplications will become n*m*k and total additions will be n x m x (k-1).

```python
list1 = [[1,2,3],[1,0,1]]
list2 = [[1,2],[2,3],[3,0]]

print(f"matrix 1 : \n{list1}")
print(f"matrix  : \n{list2}")

product = []
row = len(list1)
column = len(list2[0])
for i in range(row):
    product.append([])
    for j in range(column):
        value=0
        for k in range(3):
            value = value + list1[i][k]*list2[k][j]
        product[i].append(value)
print(f"Product matrix without using numpy : {product}\n")


# =>  Matrix Multiplication using Numpy :

print("=>  Matrix Multiplication using Numpy :\n")
import numpy as np
m1 = np.array([[1,2,3],[1,0,1]])
m2 = np.array([[1,2],[2,3],[3,0]])
print(f"matrix 1 : \n{m1}")
print(f"matrix 2 : \n{m2}")
product_by_numpy = m1.dot(m2)
print(f"Product by numpy : \n{product_by_numpy}")
```

```python

import numpy as np

def median_m1(list1,n):
    l = list1.copy()
    record = [0]*n
    max_el=float(l[0])
    max_idx=0
    k=0
    count=0
    while ((n%2==0 and count <= (n/2)) or (n%2!=0 and count <= (n+1)/2)):
        for i in range(k,n):
            if record[max_idx]!=1 and float(l[i])>= max_el:
                max_el=float(l[i])
                max_idx=i
        record[i] = 1                
        count+=1
        k+=1
    mid=max_idx
    if n%2==0:
        return (float((l[int(mid/2)]+l[int(mid/2 + 1)])/2))
    else:
        return (float(l[int((mid+1)/2)]))
    
def median_m2(list1,n):
    l = list1.copy()
    l = sorted(l)
    print(f"Sorted list : {l}")
    
    if n%2==0:
        mid1=n/2
        mid2=n/2 - 1
        median=float((int(l[int(mid1)]) + int(l[int(mid2)]))/2)
        
    else:
        median=float(int(l[int((n-1)/2)]))

    return median
    
            
n = int(input("Enter list size : "))
l1 = list()
print("\nEnter data items : \n")
for i in range(n):
    val=float(input(f"Enter data-item {i+1}: "))
    l1.append(val)
print(f"\nData items : {l1}")
median_value1 = median_m1(l1,n)
print("\n Median by method-1 is : " + str(median_value1))

median_value2 = median_m2(l1,n)
print("\n Median by method-2 is : " + str(median_value2))

print("\n Finding median by numpy : ")
l2=np.array(l1)
print(f"\n Median by numpy is : {np.median(l2)}")
```

Gradient of $f(x,y) = x^2y + y^3sin(x)$ with respect to $x = \frac{\partial f}{\partial x} = 2xy + y^3cos(x)$ .\

# Q5

```python
import jax
import jax.numpy as jnp
import math

def grad_by_m1(x,y):
    grad_x=2*x*y + y*y*y*math.cos(x)
    grad_y=x*x + 3*y*y*math.sin(x)
    return(grad_x,grad_y)

def f(x,y):
    return x*x*y + y*y*y*jnp.sin(x)

x=float(input("Enter x : "))
y=float(input("Enter y : "))

grad_f=jax.grad(f,argnums=(0,1))

Grad_x,Grad_y=grad_f(x,y)

print(f"Gradient(x,y) from traditional approach : {grad_by_m1(x,y)}")
print(f"Gradient(x,y) using JAX : ({Grad_x},{Grad_y})")
    
```

# Q6

```python
import sympy as smp
import math

x = smp.Symbol('x')
y = smp.Symbol('y')

def f(x,y):
    return x*x*y + y*y*y*smp.sin(x)

def grad_by_m1(x,y):
    grad_x=2*x*y + y*y*y*math.cos(x)
    grad_y=x*x + 3*y*y*math.sin(x)
    return grad_x,grad_y

x_val=float(input("Enter the value of x: "))
y_val=float(input("Enter the value of y: "))

analytic_grad = grad_by_m1(x_val,y_val)

print(f"The value of  gradient(x,y) analytically is : {analytic_grad}")

grad_wrt_x = smp.diff(f(x,y),x)
grad_wrt_y = smp.diff(f(x,y),y)

grad_val_x = grad_wrt_x.evalf(subs={x : x_val,y : y_val})
grad_val_y = grad_wrt_y.evalf(subs={x : x_val,y : y_val})

print(f"The value of  gradient(x,y) using sympy is : ({grad_val_x},{grad_val_y})")
```

# Q12 

Broadcasting is a way in which two numpy arrays of different dimensions can be operated upon by various arithmetic operations like addition, multiplication and so on. The array which lags in dimensions is stretched or broadcasted in other directions so that they can be operated just like two same dimensional arrays. 


```python
# CODE :

import numpy as np

m1=np.ones((2,3))

m2=np.arange(3)

# m2 lags behind in dimensions so left end is given 1 row.

# m1 (2,3) and m2 (1,3) 

# Now m2 has 1st dimension mismatched with m1 so it is stretched or broadcasted to form same dimension. therefore now  m1(2x3) =[[1,1,1,],[1,1,1]]
# and m2(2x3) =[[0,1,2],[0,1,2]]. If we add both of them we get [[1,2,3],[1,2,3]], a 2x3 matrix.

print(m1+m2)

```

# Q13

```python
import numpy as np

def min_idx_calc(arr):
    min_val = arr[0]
    min_idx=0

# Minimum element's index without using argmin:

    for i in range(len(arr)):
        if arr[i]<min_val:
            min_val=arr[i]
            min_idx=i
            
    return min_idx

arr = np.array([1,3,-2,4,0,-3])

print("Minimum element's index without using argmin: "+str(min_idx_calc(arr)))

print("Minimum element's index using argmin : "+ str(np.argmin(arr)))
```
